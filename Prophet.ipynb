```python
import pandas as pd 
from fbprophet import Prophet
```

```python
data=pd.read_csv('2004-2019.tsv',sep='\t')
```

```python
data.head(10)
```

```python
type(data['DATA INICIAL'])
```

```python
TS=pd.to_datetime(data['DATA INICIAL'])
```

```python
type(TS[0])
```

```python
df3=data['PREÇO MÉDIO REVENDA']
```

```python
df3.name='y'
```

```python
df3.head()
```

```python
df=pd.DataFrame(df3)
```

```python
df['ds']=TS
```

```python
df.head()
```

```python
clf=Prophet()
```

```python
clf.fit(df)
```

```python
future = clf.make_future_dataframe(periods=15)
future.tail()
```

```python
forecast = clf.predict(future)
```
